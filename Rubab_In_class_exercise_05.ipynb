Basic Pre-processing

```python
#Removing Punctuation
d['hashtag_tweets'] = d['hashtag_tweets'].str.replace('[^\w\s]','')
d[['hashtag_tweets','characters','hashtags','stopwords','word_count']].head()
```

```python
#Lower Casing
d['hashtag_tweets'] = d['hashtag_tweets'].apply(lambda x: " ".join(x.lower() for x in x.split()))
d[['hashtag_tweets','characters','hashtags','stopwords','word_count']].head()
```

```python
#Removing stop words
from nltk.corpus import stopwords
stop = stopwords.words('english')
d['hashtag_tweets'] = d['hashtag_tweets'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
d['hashtag_tweets'].head()

```

```python
#Removing Numerics
d['hashtag_tweets'] = d['hashtag_tweets'].apply(lambda x: ''.join([i for i in x if not i.isdigit()]))
d
```

```python
#Removing Special Characters
import re
d['hashtag_tweets'] =d['hashtag_tweets'].apply(lambda x: ''.join(re.sub(r"[^a-zA-Z0-9]+", ' ', charctr) for charctr in x ))
d
```

```python
#Stemming
from nltk.stem import PorterStemmer
st = PorterStemmer()
d['hashtag_tweets'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
```

```python
#Lemmatization
from textblob import Word
import nltk
nltk.download('wordnet')

d['hashtag_tweets'] = d['hashtag_tweets'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
d
```

POS Tagging & Features

```python
from nltk.tokenize import word_tokenize
pos = []
for sentence in d['hashtag_tweets']:
  text = word_tokenize(sentence)
  pos.append(nltk.pos_tag(text))
pos
```

```python
Adjective = []
Adverb = []
CordinatingConjunction = []
SubordinatingConjuction = []
Noun = []
Verb = []
predeterminer = []
Determiner = []

for values in pos:
  AdjectiveCount = 0
  AdverbCount = 0
  CordinatingConjunctionCount = 0
  SubordinatingConjuctionCount = 0
  NounCount = 0
  VerbCount = 0
  predeterminerCount = 0
  DeterminerCount = 0
  for word,tag in values:
    if tag == 'JJ':
      AdjectiveCount = AdjectiveCount + 1
    elif tag == 'RB':
      AdverbCount = AdverbCount + 1
    elif tag == 'CC':
      CordinatingConjunctionCount = CordinatingConjunctionCount + 1
    elif tag == 'NN':
      NounCount = NounCount + 1
    elif tag == 'VR':
      VerbCount = VerbCount + 1
    elif tag == 'PDT':
      predeterminerCount = predeterminerCount + 1
    elif tag == 'DT':
      DeterminerCount = DeterminerCount + 1
    elif tag == 'IN':
      SubordinatingConjuctionCount = SubordinatingConjuctionCount + 1
  Adjective.append(AdjectiveCount)
  Adverb.append(AdverbCount)
  CordinatingConjunction.append(CordinatingConjunctionCount)
  Noun.append(NounCount)
  Verb.append(VerbCount)
  predeterminer.append(predeterminerCount)
  Determiner.append(DeterminerCount)
  SubordinatingConjuction.append(SubordinatingConjuctionCount)
```

```python
d['No. of Adjectives'] = Adjective
d['No. of Adverbs'] = Adverb
d['No. of Cordinating Conjunctions'] = CordinatingConjunction
d['No. of Nouns'] = Noun
d['No. of Verbs'] = Verb
d['No. of Predeterminers'] = predeterminer
d['No. of of Determiners'] = Determiner
d['No. of Subordinating Conjuctions'] = SubordinatingConjuction
d
```

```python
#Inverse Document Frequency -IDF
import numpy as np

for i,word in enumerate(tf1['words']):
  tf1.loc[i, 'idf'] = np.log(d.shape[0]/(len(d[d['hashtag_tweets'].str.contains(word)])))

tf1
```

```python
#Term Frequency - Inverse Document Frequency(TF_IDF)
tf1['tfidf'] = tf1['tf'] * tf1['idf']
tf1
```
